#Langkah pertama Import Library

```python
#import library numpy
import numpy as np
#import library pandas
import pandas as pd
```

```python
#membaca file dataset
df = pd.read_csv("/content/Property Listing Dataset.csv")

#menampilkan data
df
```

#Langkah kedua check values disetiap columns dan jika ada column yang kosong isilah dengan N/A

```python
#cek values setiap column dan menampilkan jumlah values yang kosong
df.isnull().sum()
```

```python
#mengisi column yang kosong dengan N/A
df = df.fillna('N/A')

#menampilkan data
df
```

#Langkah ketiga check data duplikasi pada setiap column dan hapus salah satu

```python
#mengecek data duplikasi
df.duplicated().sum()
```

```python
#Menghapus salah satu data yang duplikasi
df = df.drop_duplicates(keep = 'last')
```

```python
#mengecek data duplikasi
df.duplicated().sum()
```

```python
#mengisi column yang kosong dengan N/A
df = df.fillna('N/A')
```

#Langkah Keempat jika da value pada column Rooms,Bathrooms,Car Parks tidak menggunakan data number maka dihapus

```python
#mengecek setiap kolom apakah ada data yang bukan number
kolom_khusus = ['Rooms','Bathrooms','Car Parks']

#perulangan untuk mengecek setiap kolom
for kolom in kolom_khusus :
  try:
    pd.to_numeric(df[kolom])
    print("Semua nilai dalam kolom '{}' adalah angka.".format(kolom))
  except ValueError:
    print("ada nilai non-Number dalam kolom".format(kolom))
```

```python
#kolom yang ingin di cek
kolom_khusus = ['Rooms','Bathrooms','Car Parks']

#perulangan untuk menghapus data non numeric
for kolom in kolom_khusus:
  #konversi nilai nilai dalam kolom ke numeric
  df[kolom]= pd.to_numeric(df[kolom],errors='coerce')

#jika gagal konversi maka hapus
df = df.dropna()

#menyimpan data yang sudah dikonversi berhasil simpan ke dalam file baru
df.to_csv('Property_Listing_Dataset_clear.csv',index=False)
```
